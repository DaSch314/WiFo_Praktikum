# Statistics

## Reading Data

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Read Excel file (adjust file name if necessary)
# If the decimal separator in the file is a comma, you can use decimal=','
df = pd.read_excel("/Users/david/Desktop/WiFo_Praktikum/Concentration_Datalist_drc.xlsx")  # , decimal=',' if needed

# Standardize column names (adjust according to your actual column names)
df.columns = ["Sample ID", "Age", "PARASITEMIA", "MSP1-IgA-Concentration", "Gender"]

# 1. Replace "n.d." in all relevant columns with NaN
#    (both "n.d" and "n.d." are captured)
df = df.replace(["n.d", "n.d."], np.nan)

# 2. Convert concentration values to numeric
#    (already handled partly by replace -> NaN, but done again for safety)
df["MSP1-IgA-Concentration"] = pd.to_numeric(df["MSP1-IgA-Concentration"], errors="coerce")

# 3. Parasitemia: replace comma with dot, then convert to numeric
#    Since NaN values already exist, we only need to process the non-NaN cells
#    A robust method:
df["PARASITEMIA"] = df["PARASITEMIA"].astype(str).str.replace(",", ".").replace("nan", np.nan)
df["PARASITEMIA"] = pd.to_numeric(df["PARASITEMIA"], errors="coerce")

# 4. Age: convert to numeric
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

# 5. Gender: convert to categorical variable (NaN values remain)
df["Gender"] = df["Gender"].astype("category")

# Overview of the dataset and missing values
print("Data types and memory usage:")
print(df.info())

print("\nFirst 5 rows:")
print(df.head())

print("\nNumber of missing values per column:")
print(df.isnull().sum())
```

## Samples by Age and Gender 

```python
# Log transformation of parasitemia (x-axis for scatter plots)
df["log_PARASITEMIA"] = np.log10(df["PARASITEMIA"] + 1)

# Create age groups (e.g., children < 18 years)
df["Age_group"] = pd.cut(df["Age"], bins=[0, 18, 100], labels=["child", "adult"])

# Binary variable for IgA detectability (True = measurable, False = not detectable)
df["IgA_detectable"] = df["MSP1-IgA-Concentration"].notna()

# Encode gender as a binary variable (for correlation matrix if needed)
gender_mapping = {"m": 0, "M": 0, "f": 1, "F": 1}  # adjust according to your actual values
df["Gender_code"] = df["Gender"].map(gender_mapping)

# Intermediate overview
print("Number of samples with detectable IgA:", df["IgA_detectable"].sum())
print("Distribution of age groups:\n", df["Age_group"].value_counts(dropna=False))
print("Gender distribution:\n", df["Gender"].value_counts(dropna=False))
```

## IgA by gender and age group

```python
# Only samples with measurable IgA
df_messbar = df.dropna(subset=["MSP1-IgA-Concentration"])

# Grouped statistics by gender
print("\n--- IgA concentration by gender ---")
print(df_messbar.groupby("Gender")["MSP1-IgA-Concentration"].describe())

# By age group
print("\n--- IgA concentration by age group ---")
print(df_messbar.groupby("Age_group")["MSP1-IgA-Concentration"].describe())

# Detectability by gender
print("\n--- IgA detectability by gender ---")
print(pd.crosstab(df["Gender"], df["IgA_detectable"], margins=True))
```

## Heatmap correlation of all variables

Spearman because is doesnt asume normal distribution and is robust against outliers (because of ranks)

```python
# Select numeric columns
numeric_cols = ["MSP1-IgA-Concentration", "log_PARASITEMIA", "Age", "Gender_code"]

# Use only complete cases (no NaN values)
corr_data = df[numeric_cols].dropna()

# Spearman correlation (robust to outliers)
corr_matrix = corr_data.corr(method="spearman")

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            vmin=-1, vmax=1, square=True, linewidths=0.5)

plt.title("Spearman correlations (n.d. samples excluded)")
plt.tight_layout()
plt.show()
```

## Mann-Whitney-U-Test

non parametric Mann-Whitnes-U-test because data is not 
- normally distributed
- test robiust to outliers (compares rank of values)
- suitable for comparuing two indepoendet groups 


```python
# Gruppen extrahieren (nur messbare Werte, beide Schreibweisen)
gruppe_m = df_messbar[df_messbar["Gender"].isin(["m", "M"])]["MSP1-IgA-Concentration"]
gruppe_f = df_messbar[df_messbar["Gender"].isin(["f", "F"])]["MSP1-IgA-Concentration"]

stat, p = stats.mannwhitneyu(gruppe_m, gruppe_f, alternative="two-sided")
print(f"Mann-Whitney U-Test für IgA nach Geschlecht: p = {p:.4f}")
```

Result shows, that even though a trend in higher IgA values in males was observed it is not statistically rsignificant if you choose standard significance level of 0.05.

p = 0,092 > 0.05

## Chi-Squared-Test of IgA and Gender

to test if there is a difference if there is a significant difference in of how often IgA is detected in males vs females

```python
# Contingency table (excluding missing gender values)
contingency = pd.crosstab(df["Gender"].dropna(), df["IgA_detectable"].loc[df["Gender"].notna()])

print("Contingency table:")
print(contingency)

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print(f"Chi-square test: p = {p:.4f}")
```

Females: 40/59 (67.8%) IgA detectable  
Males: 55/68 (80.9%) IgA detectable

p = 13.64 > 0.05

--> not statistically significant

```python
# Daten für Korrelation
korr_data = df_messbar[["PARASITEMIA", "MSP1-IgA-Concentration"]].dropna()
rho, p = stats.spearmanr(korr_data["PARASITEMIA"], korr_data["MSP1-IgA-Concentration"])
print(f"Spearman-Korrelation Parasitämie vs. IgA: ρ = {rho:.3f}, p = {p:.4f}")
```
