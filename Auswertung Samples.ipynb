```python
import matplotlib.pyplot as plt
import numpy as np

# Absorbance-Werte (jeweils Duplikate, Plate 2)
absorbance_1_plate1 = [1.472, 0.908, 0.572, 0.324, 0.185, 0.106, 0.074, 0.059]
absorbance_2_plate1 = [1.446, 0.968, 0.557, 0.311, 0.182, 0.108, 0.076, 0.056]

# Mittelwert der Messungen
absorbance_mean_plate1 = [(a + b) / 2 for a, b in zip(absorbance_1_plate1, absorbance_2_plate1)]

# Blank-Werte (jeweils Duplikate, Plate 2)
blanks_1_plate1= [0.043, 0.048, 0.045, 0.047, 0.046, 0.044]
blanks_2_plate1 = [0.044, 0.045, 0.049, 0.045, 0.09, 0.043]

# Mittelwert der Blanks
blanks_mean_plate1 = [(a + b) / 2 for a, b in zip(blanks_1_plate1, blanks_2_plate1)]
blank_mean_plate1 = np.mean(blanks_mean_plate1)

# Hintergrundkorrektur
absorbance_corrected_plate1 = [value - blank_mean_plate1 for value in absorbance_mean_plate1]

# Konzentrationen
concentration = [100, 50, 25, 12.5, 6.25, 3.125, 1.5625, 0.78125]

# Lineare Regression
coeffs_plate1 = np.polyfit(concentration, absorbance_corrected_plate1, 1)
regression_line_plate1 = np.polyval(coeffs_plate1, concentration)

# Plot
plt.plot(concentration, absorbance_corrected_plate1, 'o', label='Korrigierte Messpunkte')
plt.plot(concentration, regression_line_plate1, '-', label='Lineare Regression')
plt.xlabel('Konzentration (µg/dl)')
plt.ylabel('Absorbance(minus background)')
plt.title('IgA Standardkurve (Plate 1)')
plt.grid(True)
plt.legend()
plt.show()








import matplotlib.pyplot as plt
import numpy as np

# Absorbance-Werte (jeweils Duplikate)
absorbance_1_plate2 = [1.357, 0.901, 0.561, 0.332, 0.181, 0.111, 0.077, 0.058]
absorbance_2_plate2 = [1.43, 0.905, 0.554, 0.328, 0.184, 0.11, 0.074, 0.058]

# Mittelwert der Messungen
absorbance_mean_plate2 = [(a + b) / 2 for a, b in zip(absorbance_1_plate2, absorbance_2_plate2)]

# Blank-Werte (12 Messungen)
blanks_1_plate2 = [0.044, 0.044, 0.045, 0.046, 0.043, 0.044, 0.045, 0.044, 0.056, 0.047, 0.046, 0.045]
blanks_2_plate2 = [0.044, 0.045, 0.049, 0.045, 0.09, 0.043]

# Mittelwert der Blanks
blanks_mean_plate2 = [(a + b) / 2 for a, b in zip(blanks_1_plate2, blanks_2_plate2)]
blank_mean_plate2 = np.mean(blanks_mean_plate2)

# Hintergrundkorrektur
absorbance_corrected_plate2 = [value - blank_mean_plate2 for value in absorbance_mean_plate2]

# Konzentrationen
concentration = [100, 50, 25, 12.5, 6.25, 3.125, 1.5625, 0.78125]

# Lineare Regression
coeffs_plate2 = np.polyfit(concentration, absorbance_corrected_plate2, 1)
regression_line_plate2 = np.polyval(coeffs_plate2, concentration)

# Plot
plt.plot(concentration, absorbance_corrected_plate2, 'o', label='Korrigierte Messpunkte')
plt.plot(concentration, regression_line_plate2, '-', label='Lineare Regression')
plt.xlabel('Konzentration (µg/dl)')
plt.ylabel('Absorbance(minus background)')
plt.title('IgA Standardkurve (Plate 2)')
plt.grid(True)
plt.legend()
plt.show()



```

```python
import numpy as np

# Gemessene Probenwerte (jeweils Duplikate)
samples_plate1 = [
    [0.052, 0.052],
    [0.044, 0.044],
    [0.071, 0.06],
    [0.094, 0.093],
    [0.046, 0.05],
    [0.053, 0.068],
    [0.128, 0.087],
    [0.044, 0.054],
    [0.095, 0.092],
    [0.061, 0.061],
    [0.065, 0.062],
    [0.075, 0.065],
    [0.076, 0.053],
    [0.05, 0.045],
    [0.06, 0.062],
    [0.071, 0.071]
]

# Mittelwert pro Probe
samples_mean_plate1 = [np.mean(pair) for pair in samples_plate1]

# Blank-Mittelwert von Plate 1 (aus deinem Code)
blank_mean_plate1 = 0.05208333333333333  # ggf. anpassen, falls sich der Wert geändert hat

# Hintergrundkorrektur
samples_corrected_plate1 = [value - blank_mean_plate1 for value in samples_mean_plate1]

# Regressionskoeffizienten von Plate 1 (aus deinem Modell)
coeffs_plate1 = np.polyfit(
    [100, 50, 25, 12.5, 6.25, 3.125, 1.5625, 0.78125],
    [value - blank_mean_plate1 for value in [
        (a + b) / 2 for a, b in zip(
            [1.472, 0.908, 0.572, 0.324, 0.185, 0.106, 0.074, 0.059],
            [1.446, 0.968, 0.557, 0.311, 0.182, 0.108, 0.076, 0.056]
        )
    ]],
    1
)

# Berechnung der Konzentration jeder Probe (Umkehrung der Geraden: y = m*x + b -> x = (y-b)/m)
m, b = coeffs_plate1
expected_concentration_plate1 = [(y - b) / m for y in samples_corrected_plate1]

# Ausgabe
for i, (mean, corr, conc) in enumerate(zip(samples_mean_plate1, samples_corrected_plate1, expected_concentration_plate1), 1):
    print(f"Probe {i}: Mittelwert = {mean:.3f}, korrigiert = {corr:.3f}, Konzentration = {conc:.2f} µg/dl")
```
